```python
import pandas as pd
import json
```

```python
df = pd.read_excel('dataset.xlsx', sheet_name='Sheet1')
df
```

```python
# 新建一个df，列为df中的Institution，Country，Contributor，Samples列
df1 = df[['Institution', 'Country', 'Contributor', 'Samples']]
df1
```

```python
# 将Contributor列中同样的作者合并，Samples列中的样本数相加
df2 = df1.groupby(['Institution', 'Country', 'Contributor']).sum()

# 将df2中的index转换为columns
df2 = df2.reset_index()

df2
```

```python
# df2转为字典
df2_dict = df2.to_dict(orient='records')
df2_dict
```

```python
output = {}

for item in df2_dict:
    if item['Country'] not in output:
        output[item['Country']] = {}
    if item['Contributor'] not in output[item['Country']]:
        output[item['Country']][item['Contributor']] = {}
        output[item['Country']][item['Contributor']]['Unit'] = item['Institution']
        output[item['Country']][item['Contributor']]['Samples'] = str(item['Samples'])

# 将output字典转换为json格式
output_json = json.dumps(output, ensure_ascii=False, indent=4)

# 将output_json写入文件
with open('map.json', 'w', encoding='utf-8') as f:
    f.write(output_json)
```
