```python
import pandas as pd


df_sample = pd.read_csv('sample.tsv', sep='\t')
df_sample
```

```python
# 删除Sample Name这一列中的Myceliophthora_thermophila_
df_sample['Sample Name'] = df_sample['Sample Name'].str.replace('Myceliophthora_thermophila_', '')
df_sample['Sample Name'] = df_sample['Sample Name'].str.replace('Myceliphothora thermophila ', '')


# 将Species这一列Thermothelomyces thermophilus改为Myceliophthora thermophila
df_sample['Species'] = df_sample['Species'].str.replace('Thermothelomyces thermophilus ', 'Myceliphothora thermophila')
```

```python
# df_sample中，Sample Number这一列有重复值，将重复值的行合并,其中Study Title这一列用；分开
df_sample = df_sample.groupby('Sample Number').agg({
    'Sample ID':lambda x: ','.join(x),
    'Run':lambda x: ','.join(x),
    'Series ID':lambda x: ','.join(x),
    'Study Title':lambda x: ';'.join(x),
    'Species':lambda x: ','.join(x),
    'Strain':lambda x: ','.join(x),
    'Model':lambda x: ','.join(x),
    'Library Strategy':lambda x: ','.join(x),
    'Sample Name':lambda x: ','.join(x),
    'Carbon Sources':lambda x: ','.join(x),
    'Temperature':lambda x: ','.join(x),

    }).reset_index()

df_sample
```

```python
# Series ID这一列中，存在多个值，如果值相同则值合并
df_sample['Series ID'] = df_sample['Series ID'].str.split(',').str[0]
df_sample['Study Title'] = df_sample['Study Title'].str.split(';').str[0]
df_sample['Species'] = df_sample['Species'].str.split(',').str[0]
df_sample['Strain'] = df_sample['Strain'].str.split(',').str[0]
df_sample['Model'] = df_sample['Model'].str.split(',').str[0]
df_sample['Library Strategy'] = df_sample['Library Strategy'].str.split(',').str[0]
df_sample['Carbon Sources'] = df_sample['Carbon Sources'].str.split(',').str[0]
df_sample['Temperature'] = df_sample['Temperature'].str.split(',').str[0]

df_sample
```

```python
# 保存
df_sample.to_csv('sample_fname.tsv', sep='\t', index=False)
```
