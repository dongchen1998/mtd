```python
import pandas as pd


df_sample = pd.read_csv('sample.tsv', sep='\t')
df_sample
```

```python
# 删除Sample Name这一列中的Myceliophthora_thermophila_
df_sample['Sample Name'] = df_sample['Sample Name'].str.replace('Myceliophthora_thermophila_', '')
df_sample['Sample Name'] = df_sample['Sample Name'].str.replace('Myceliphothora thermophila ', '')


# 将Species这一列Thermothelomyces thermophilus改为Myceliophthora thermophila
df_sample['Species'] = df_sample['Species'].str.replace('Thermothelomyces thermophilus ', 'Myceliphothora thermophila')
```

```python
# 保存
df_sample.to_csv('sample.tsv', sep='\t', index=False)
```
