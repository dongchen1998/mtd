```python
import pandas as pd

# mtd/search/myceliophthora_thermophila/data_processing_file/Symbol_id.csv
df = pd.read_csv('Symbol_id.csv')
df.head()
```

```python
# 遍历每一行，如果Gene_Symbol有重复，就输出
for i in range(len(df)):
    for j in range(i+1, len(df)):
        if df.loc[i, 'Gene_symbol'] == df.loc[j, 'Gene_symbol']:
            print(df.loc[i, 'Gene_symbol'])
```
