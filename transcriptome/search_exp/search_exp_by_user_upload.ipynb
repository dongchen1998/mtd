```python
import pandas as pd
import os
import json

"""
场景假设：
1. 用户提上传基因list，格式有逗号分隔与换行符分隔两种
2. 根据基因list，返回相关表达谱
"""

def gene_exp_df_huanhuang(gene_list_path, exp_path):
    """
    Description: 根据用户提交的基因list，找到对应的基因的表达量(GSM)
    Arguments:
        gene_list: 用户提交的基因list
    Returns:
        df: 以GSM ID为表头的表达谱的dataframe
    """
    with open(gene_list_path, 'r') as f:
        gene_list = [line.strip() for line in f.readlines()]

    gene_list = [gene.upper() for gene in gene_list]

    df_exp = pd.read_csv(exp_path, sep='\t')
    df = df_exp[df_exp['Gene id'].isin(gene_list)]

    return df


def gene_exp_df_fname_huanhuan(gene_list_path, exp_fname_path):
    """
    Description: 根据用户提交的基因list，找到对应的基因的表达量(F)
    Arguments:
        gene_list: 用户提交的基因list
    Returns:
        df: 以GSM ID为表头的表达谱的dataframe
    """
    with open(gene_list_path, 'r') as f:
        gene_list = [line.strip() for line in f.readlines()]
        
    gene_list = [gene.upper() for gene in gene_list]

    df_exp_fname = pd.read_csv(exp_fname_path, sep='\t')
    df = df_exp_fname[df_exp_fname['Gene id'].isin(gene_list)]

    return df


def gene_exp_df_douhao(gene_list, exp_path):
    """
    Description: 处理用户提交的基因list(txt)，以逗号分隔，根据df_exp，找到对应的基因的表达量(GSM)
    Arguments:
        gene_list: 用户提交的基因list
    Returns:
        df: 以GSM ID为表头的表达谱的dataframe
    """
    with open(gene_list, 'r') as f:
        content = f.read()

    gene_list = content.replace(',', '\n').split()
    gene_list = [gene.upper() for gene in gene_list]

    df_exp = pd.read_csv(exp_path, sep='\t')
    df = df_exp[df_exp['Gene id'].isin(gene_list)]

    return df


def gene_exp_df_fname_douhao(gene_list, exp_fname_path):
    """
    Description: 处理用户提交的基因list(txt)，以逗号分隔，根据df_exp_fname，找到对应的基因的表达量(F)
    Arguments:
        gene_list: 用户提交的基因list
    Returns:
        df: 以GSM ID为表头的表达谱的dataframe
    """
    with open(gene_list, 'r') as f:
        content = f.read()

    gene_list = content.replace(',', '\n').split()
    gene_list = [gene.upper() for gene in gene_list]

    df_exp_fname = pd.read_csv(exp_fname_path, sep='\t')
    df = df_exp_fname[df_exp_fname['Gene id'].isin(gene_list)]

    return df
```
