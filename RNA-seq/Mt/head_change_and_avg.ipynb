```python
mt_exp = pd.read_csv("Mt_TPM_Run221_gsm.csv")
mt_sample = pd.read_csv("sample.tsv", sep="\t")

id_map = dict(zip(mt_sample["Sample Name"], mt_sample["Sample ID"])) 

new_columns = ["Gene id"] + [id_map[col] for col in mt_exp.columns[1:]] 
mt_exp.columns = new_columns

df_mtd_id = pd.DataFrame(mt_exp)
```

```python
df_mtd_id
```

```python
import pandas as pd
import copy

df_mtd_id.columns = df_mtd_id.columns.str.replace('\.\d', '')

# copy.deepcopy()函数用于创建原始对象的完全独立副本（包括嵌套对象和所有内部对象），以便在进行更改时不会影响原始对象
df_mtd_id_new = copy.deepcopy(df_mtd_id.iloc[:,0:1])
columns = df_mtd_id.iloc[:,1:].columns

for col in columns:
    if len(pd.DataFrame(df_mtd_id[col]).columns)>=2: # series转DF
        df_mtd_id_new[col] = df_mtd_id[col].mean(axis=1)
    else:
        df_mtd_id_new[col] = df_mtd_id[col]


df_mtd_id_new = df_mtd_id_new.round(2)
df_mtd_id_new

```

```python
df_mtd_id_new.to_csv('exp.csv',index=False)
```
