```python
import pandas as pd

# 读取CSV文件，假设文件名为data.csv，第一列为基因ID，其余列为样本编号
df = pd.read_csv('./Mt/Mt_FPKM_173Run.csv', index_col=0)

# 读取文本文件，假设文件名为gene_list.txt，每行包含一个基因ID号码
with open('./Mt/Mt_autophagy_list_KEGG.txt', 'r') as f:
    gene_list = [line.strip() for line in f]

# 选择特定基因的数据，将结果保存到新的DataFrame对象中
df_selected = df.loc[gene_list, :]

# 将结果保存为CSV文件，假设文件名为result.csv
df_selected
df_selected.to_csv('./Mt/Mt_autophagy_list_KEGG.csv')
```
