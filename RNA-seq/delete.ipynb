```python
with open('./Nc/NC_donwnload.txt', 'r') as file:
    lines = file.readlines()

numbers = []
for line in lines:
    numbers.extend(line.split())

unique_numbers = list(set(numbers))

with open('./Nc/NC_donwnload.txt', 'w') as file:
    for number in unique_numbers:
        file.write(number + '\n')
```
